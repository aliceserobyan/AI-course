```python
avg_price = df["price_usd"].mean()
print(f"The average car price is: ${avg_price}")
```

```python
powerful_cars = df[df["hp"] > 300]
powerful_cars
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(df["model"], df["price_usd"])
plt.title("Price of Car Models")
plt.ylabel("Price (USD)")
plt.xlabel("Model")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df["hp"], df["price_usd"])

plt.title("Horsepower vs. Price")
plt.xlabel("Horsepower (HP)")
plt.ylabel("Price (USD)")

plt.show()
```
