```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
# Load the Titanic dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
# 1. Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# 2. Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# 3. Drop Cabin column
df.drop(columns=['Cabin'], inplace=True)

# 4. Cap Fare at the 99th percentile to reduce outliers
q99 = df['Fare'].quantile(0.99)
df['Fare'] = df['Fare'].clip(upper=q99)

df.head()
```

```python
# One-hot encode Sex and Embarked (nominal data)
df_ohe = pd.get_dummies(df[['Sex', 'Embarked']], drop_first=True)

# Show new encoded columns
df_ohe.head()
```

```python
# Label Encoding for Pclass (ordinal)
le = LabelEncoder()

df['Pclass'] = le.fit_transform(df['Pclass'])

df['Pclass'].head()
```

```python
df['Pclass'].value_counts()
```

```python
# Select numeric columns
numeric_cols = df[['Age', 'Fare']]

# StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(numeric_cols)

# Convert back to DataFrame
df_scaled = pd.DataFrame(scaled, columns=['Age_scaled', 'Fare_scaled'])
df_scaled.describe()
```

```python
# MinMaxScaler
mm = MinMaxScaler()
normalized = mm.fit_transform(numeric_cols)

df_norm = pd.DataFrame(normalized, columns=['Age_norm', 'Fare_norm'])
df_norm.describe()
```
